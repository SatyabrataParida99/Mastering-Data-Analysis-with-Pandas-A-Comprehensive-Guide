**Steps to Perform Frequency Analysis on Categorical Columns**
   - **Create Data** – Define a categorical column in a DataFrame.
   - **Compute Frequency Table** – Manually count occurrences for each category.
   - **Use value_counts()** – Generate category counts efficiently.
   - **Interpret Output** – value_counts() returns a Pandas Series with counts. 

```python
continent_vc=visa_df['continent'].value_counts()
keys=continent_vc.keys()
values=continent_vc.values
pd.DataFrame(zip(keys,values),
             columns=['continent','No of Applicants'])
```

**Statistical Summary of prevailing_wage Column**

- Count: Total number of observations in the dataset.
    
- Minimum (Min): The lowest recorded prevailing wage.

- Mean: The average prevailing wage across all observations.
    
- Median (50th Percentile): The middle value when wages are sorted.

- 25th Percentile (25p): The wage value below which 25% of observations fall.

- 50th Percentile (50p): Same as the median value.

- 75th Percentile (75p): The wage value below which 75% of observations fall.
    
- Maximum (Max): The highest recorded prevailing wage.

**Emperical Rule**

- The Empirical Rule (also known as the 68-95-99.7 Rule) is a statistical guideline for normal distributions. It states that:

- 🔹 68% of data falls within 1 standard deviation (σ) of the mean (μ).

- 🔹 95% of data falls within 2 standard deviations (σ) of the mean (μ).

- 🔹 99.7% of data falls within 3 standard deviations (σ) of the mean (μ).

<img src="https://i0.wp.com/statisticsbyjim.com/wp-content/uploads/2021/08/empirical_rule_graph2.png?fit=572%2C384&amp;ssl=1" jsaction="" class="sFlh5c FyHeAf iPVvYb" style="max-width: 572px; height: 203px; margin: 0px; width: 302px;" alt="Empirical Rule: Definition &amp; Formula - Statistics By Jim" jsname="kn3ccd" aria-hidden="false">

```python
wage_data=visa_df['prevailing_wage']
mean=wage_data.mean()
sd=wage_data.std()
lb=mean-1*sd
ub=mean+1*sd
cond1=wage_data>lb
cond2=wage_data<ub
con=cond1&cond2
v1=len(visa_df[con])
print(v1)
v2=68*len(wage_data)/100
print(v2)
v1==v2
```

- The **Empirical Rule* applies only to **normal distributions.** However, **wage data** is often **skewed,** meaning it has **outliers** that distort the distribution.
  
- This skewness causes the Empirical Rule to fail, as the data does not follow a perfect bell-shaped curve.

**Handling Outliers in Data Analysis**

**Method 1: Dropping Outliers**
 - Outliers can be removed if they constitute less than 2% of the total data.
   
 - Not recommended as it leads to data loss.

   
 - Removing outliers also removes associated values in other columns, affecting overall analysis.

   
 - Should only be used when outliers are truly erroneous or negligible.

**Method 2: Replacing with Median**

 - Since median is unaffected by outliers, replacing them with the median is a robust approach.
   
 - Preserves data structure and avoids bias.

   
 - Suitable when outliers result from natural variations rather than errors.

**Method 3: Capping Outliers**
 - Instead of replacing outliers with the median, another approach is to cap them at specific threshold values.

```python
Q1=np.quantile(wage_data,0.25)
Q2=np.quantile(wage_data,0.50)
Q3=np.quantile(wage_data,0.75)
IQR=Q3-Q1
lb=Q1-(1.5*IQR)
ub=Q3+(1.5*IQR)
median=wage_data.median()
new_data=[]
for value in wage_data:
    if value<lb or value>ub:
        new_data.append(median)
    else:
         new_data.append(value)
visa_df['prevailing_wage']=new_data
```

**Understanding Data Skewness & Transformations**

- In data analysis, skewness refers to the asymmetry of a dataset's distribution. There are three types of skewness:

   - 1️⃣ Positive Skew – Long tail on the right (right-skewed).

   - 2️⃣ Negative Skew – Long tail on the left (left-skewed).

   - 3️⃣ No Skew – Symmetrical distribution (normal distribution).
 
**Why Does Skewness Occur?**

- 🔹 Outliers often cause skewness.
  
- 🔹 Even after treating outliers, some skewness may still remain.

  
- 🔹 Most mathematical and statistical models assume that data **follows a normal distribution.**

**How to Handle Skewness?**

- To transform skewed data into a normal distribution, we use transformation techniques:
- 1️⃣ Log Transformation
  
    - Works well for highly skewed data.
  
    - Reduces right skew by compressing large values.

- 2️⃣ Reciprocal Transformation

    - Reduces the impact of large values.
  
    - Handles right-skewed data effectively.

- 3️⃣ Square Root (Sqrt) Transformation

    - Suitable for moderate skewness.
  
    - Less aggressive than log transformation.

- 4️⃣ Exponential Transformation

     - Used when data is negatively skewed.
  
     - Helps spread out small values.

- 5️⃣ Power Transformation (Box-Cox & Yeo-Johnson)

     - Box-Cox – Used for positive values.

     - Yeo-Johnson – Works for both positive and negative values.

**Joining DataFrames in Pandas**

- In Pandas, multiple DataFrames can be combined using various methods to facilitate data analysis and manipulation.

- **Methods to Join DataFrames:**

    - merge() – Similar to SQL joins, allowing joins on specific columns.
      
    - join() – Used for joining on index by default.
 
      
    - concat() – Combines DataFrames along rows or columns.
 
      
    - Function Usage: pd.merge(df1, df2, on='column_name')
 
      
    - Similar to SQL Joins: Supports inner, outer, left, and right joins.

```python
import pandas as pd
```

```python
d1={'ID':[23,27,37],
    'Name':['Satya','Himanshu','Ranjan']}
d2={'ID':[25,36,17],
    'Age':[20,25,30]}

df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
```

```python
pd.merge(df1,df2,on='ID',how='inner')
```

```python
pd.merge(df1,df2,on='ID',how='outer')
```

```python
pd.merge(df1,df2,on='ID',how='left')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```
